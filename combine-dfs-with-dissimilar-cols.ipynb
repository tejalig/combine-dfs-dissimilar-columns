```python
import pandas as pd
import json
from pymongo import MongoClient
client = MongoClient('localhost:27017')
db = client.example
```

```python
df1 = pd.DataFrame({'FirstName' : ['fname1', 'fname2', 'fname3'], 'LastName' : ['lname1', 'lname2', 'lname3'], 'Age' : [10, 12, 13]})
```

```python
df1
```

```python
df2 = pd.DataFrame({'FirstName' : ['fname4', 'fname5', 'fname6']})
df2
```

```python
df3 = pd.DataFrame({'LastName' : ['lname7', 'lname8', 'lname9'], 'Country' : ['Country7', 'Country8', 'Country9']})
df3
```

```python
df4 = pd.DataFrame({'Job' : ['job10', 'job11', 'job12']})
df4
```

```python
pd.concat([df1, df2, df3, df4]).reset_index(drop = True)
```

```python
js_data = json.loads(df1.to_json(orient = 'records'))
js_data2 = json.loads(df2.to_json(orient = 'records'))
js_data3 = json.loads(df3.to_json(orient = 'records'))
js_data4 = json.loads(df4.to_json(orient = 'records'))
```

```python
db.example.insert(js_data)
db.example.insert(js_data2)
db.example.insert(js_data3)
db.example.insert(js_data4)
```

```python
cursor = db.example.find()
cursor = list(cursor)
combined = pd.DataFrame(cursor)
```
